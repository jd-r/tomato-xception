```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random

DATADIR = "G:\\My Drive\\Colab Notebooks\\tomato\\preprocessed_data"
CATEGORIES = ["Tomato___Bacterial_spot", "Tomato___Early_blight", "Tomato___healthy", "Tomato___Late_blight", "Tomato___Leaf_Mold", "Tomato___Septoria_leaf_spot", "Tomato___Spider_mites Two-spotted_spider_mite", "Tomato___Target_Spot", "Tomato___Tomato_mosaic_virus", "Tomato___Tomato_Yellow_Leaf_Curl_Virus"]
```

```python
def create_training_data(directory):
    training_data = []
    for category in CATEGORIES:
        path = os.path.join(DATADIR, directory)
        path = os.path.join(path, category) # path to images
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img))
                img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
                training_data.append([img_array, class_num])
            except Exception as e:
                print(e)
    random.shuffle(training_data)
    return training_data
```

```python
X_val = []
y_val = []
training_data = create_training_data('validation')
for features, label in training_data:
    X_val.append(features)
    y_val.append(label)
X_val = np.asarray(X_val, np.uint8)
y_val = np.asarray(y_val, np.uint8)
```

```python
X_test = []
y_test = []
training_data = create_training_data('test')
for features, label in training_data:
    X_test.append(features)
    y_test.append(label)
X_test = np.asarray(X_test, np.uint8)
y_test = np.asarray(y_test, np.uint8)
```

```python
X_train = []
y_train = []
training_data = create_training_data('train')
for features, label in training_data:
    X_train.append(features)
    y_train.append(label)
X_train = np.asarray(X_train, np.uint8)
y_train = np.asarray(y_train, np.uint8)
```

```python
np.savez("C:\\Users\\jdrdz\\Desktop\\tomato.npz", 
         X_train=X_train, y_train=y_train,
         X_val=X_val, y_val=y_val,
         X_test=X_test, y_test=y_test)
```
