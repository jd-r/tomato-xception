```python
import tensorflow as tf
import os, shutil
image  = tf.keras.preprocessing.image
TRAIN_SIZE = 700 # 70%
VAL_SIZE = 150   # 15%
TEST_SIZE = 150  # 15%
```

```python
original_dataset_dir = "G:\\My Drive\\Colab Notebooks\\tomato\\Plant_leave_diseases_dataset_without_augmentation"
base_dir = "G:\\My Drive\\Colab Notebooks\\tomato\\preprocessed_data"
os.mkdir(base_dir)

train_dir = os.path.join(base_dir, 'train')
os.mkdir(train_dir)
validation_dir = os.path.join(base_dir, 'validation')
os.mkdir(validation_dir)
test_dir = os.path.join(base_dir, 'test')
os.mkdir(test_dir)

CATEGORIES = ["Tomato___Bacterial_spot", "Tomato___Early_blight", "Tomato___healthy", "Tomato___Late_blight", "Tomato___Leaf_Mold", "Tomato___Septoria_leaf_spot", "Tomato___Spider_mites Two-spotted_spider_mite", "Tomato___Target_Spot", "Tomato___Tomato_mosaic_virus", "Tomato___Tomato_Yellow_Leaf_Curl_Virus"]
for categorie in CATEGORIES:
    os.mkdir(os.path.join(train_dir, categorie))
    os.mkdir(os.path.join(validation_dir, categorie))
    os.mkdir(os.path.join(test_dir, categorie))
```

```python
for categorie in CATEGORIES:
    fnames = ['image ({}).jpg'.format(i) for i in range(1, VAL_SIZE + 1)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, categorie)
        dst_dir = os.path.join(validation_dir, categorie)
        src = os.path.join(src_dir, fname)
        dst = os.path.join(dst_dir, fname)
        shutil.copyfile(src, dst)
    fnames = ['image ({}).jpg'.format(i) for i in range(VAL_SIZE + 1, TEST_SIZE + VAL_SIZE + 1)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, categorie)
        dst_dir = os.path.join(test_dir, categorie)
        src = os.path.join(src_dir, fname)
        dst = os.path.join(dst_dir, fname)
        shutil.copyfile(src, dst)
    fnames = ['image ({}).jpg'.format(i) for i in range(TEST_SIZE + VAL_SIZE + 1, TEST_SIZE + VAL_SIZE + TRAIN_SIZE + 1)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, categorie)
        dst_dir = os.path.join(train_dir, categorie)
        src = os.path.join(src_dir, fname)
        if os.path.isfile(src):
            dst = os.path.join(dst_dir, fname)
            shutil.copyfile(src, dst)
```

```python
import tensorflow as tf
import os, shutil
image  = tf.keras.preprocessing.image
from google.colab import drive
import glob

drive.mount('/content/gdrive', force_remount=True)
TRAIN_SIZE = 100
```

```python
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')
```

```python
train_dir = '/content/gdrive/My Drive/Colab Notebooks/tomato/tomato_better_dataset'
for categorie in CATEGORIES:
    images_dir = os.path.join(train_dir, categorie)
    dir_size = len(os.listdir(images_dir))
    fnames = [os.path.join(images_dir, fname) for fname in os.listdir(images_dir)]
    fnames = iter(fnames)
    
    i = 0
    max_size = TRAIN_SIZE // dir_size
    while dir_size + i < TRAIN_SIZE:
        img_path = next(fnames)
        img = image.load_img(img_path, target_size=(256, 256))
        x = image.img_to_array(img)
        x = x.reshape((1,) + x.shape)
        
        j = 0
        for batch in datagen.flow(x, save_to_dir=images_dir, save_prefix='image', save_format='jpeg', batch_size=1):
            j += 1
            i += 1
            if j == max_size:
                break
```
